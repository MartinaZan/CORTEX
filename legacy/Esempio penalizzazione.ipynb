```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def penalization(x, interval, max_val=10):
    a, b = interval
    return max_val * ((x - b) / (b - a))**2
```

```python
# 183 in (157,206)
# 2.2032486463973346
```

```python
# Funzione di penalizzazione
interval = (157,206)
x = np.linspace(interval[0], interval[1], 500)
y = penalization(x, interval)

# Esempio punto
x_point = 183
y_point = penalization(x_point, interval)

# Plot
plt.plot(x, y)
plt.scatter(x_point, y_point, color='red')
plt.axvline(interval[0], color='gray', linestyle='--')
plt.axvline(interval[1], color='gray', linestyle='--')
plt.grid(True)
plt.title('Penalization')
plt.show()
```
