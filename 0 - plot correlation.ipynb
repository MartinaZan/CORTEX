```python
from EEG_utils.EEG_utils import *
import seaborn as sb
from matplotlib.animation import FuncAnimation
from matplotlib.animation import writers
from IPython.display import HTML
```

Settings

```python
root_folder = "C:\\Users\\zanno\\physionet.org\\files\\chbmit\\1.0.0\\"
output_path = "output\plot correlation"
```

```python
num_points = 150
num_node_features = 10
lag_nodes = 1
quantile_edges = 0.25
```

```python
patient_id = "chb03"
list_record_id = ["01", "02", "03", "04", "34", "35", "36"]

list_corr_sec = [1, 3, 5, 7, 10]
```

Function for animation

```python
def animate(t, results, axs, fig, list_corr_sec):
    for ax in axs:
        ax.clear()
        
    print(t)

    for i in range(len(list_corr_sec)):
        sb.heatmap(results[list_corr_sec[i]]["corr"][t], cmap="Blues", square=True, vmin=-1, vmax=1, ax=axs[i], cbar=False)
        axs[i].set_title(f"{list_corr_sec[i]} sec")

    if results[list_corr_sec[0]]["seizure_class"][t] == 1:
        fig.suptitle("Ongoing seizure", y=0.8)
    else:
        fig.suptitle("", y=0.8)

    plt.tight_layout()
```

Plot

```python
for record_id in list_record_id:
    print(f"{patient_id}_{record_id}")

    # Create patient and compute correlations
    results = {}

    for corr_sec in list_corr_sec:
        print(f"  {corr_sec}")
        patient = Patient(FilePatient(root_folder, patient_id, record_id),
                        num_points=num_points, num_node_features=num_node_features, lag_nodes=lag_nodes,
                        quantile_edges=1, corr_sec=corr_sec)

        patient.extract_seizure_info()
        patient.load_data()
        patient.compute_indices()

        _, corr, _, _, _, seizure_class = create_graph(patient)

        results[corr_sec] = {
            'corr': corr,
            'seizure_class': seizure_class
        }

    print("    Creating video...")

    # Create animation
    fig, ax = plt.subplots(1,len(list_corr_sec), figsize=(15,5))

    ani = FuncAnimation(
        fig,
        animate,
        frames = len(results[list_corr_sec[0]]["corr"]) - 1,
        fargs = (results, ax, fig, list_corr_sec),
        repeat = False
    )

    ani.save(f"{output_path}\{patient_id}_{record_id}.mp4", writer="ffmpeg", fps=5)
    # HTML(ani.to_jshtml())
```
